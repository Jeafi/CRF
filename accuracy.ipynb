```python
import codecs
```

```python
list_t=[]
```

```python
input_data = codecs.open('Lceshiji.txt', 'r', 'utf-8')
```

```python
for line in input_data.readlines():
    if len(line) != 1:
        l = line.split('\t')[1].split('\n')[0]
        list_t.append(l)
```

```python
data = codecs.open('test_result', 'r+b', 'utf-8')
```

```python
list_r=[]
for line in data.readlines():
    try:
        l = line.split('\t')[2].split('\r')[0]
    except :
        pass
    else:
        list_r.append(l)
```
